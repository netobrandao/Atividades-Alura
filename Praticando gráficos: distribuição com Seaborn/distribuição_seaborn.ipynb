```python
#importando bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Atividade 1

```python
# Importando os dados
data = pd.read_csv('https://raw.githubusercontent.com/netobrandao/Atividades-Alura/refs/heads/main/Praticando%20gr%C3%A1ficos%3A%20distribui%C3%A7%C3%A3o%20com%20Seaborn/atividade_1.csv')

# Construindo o Grafico
plt.figure(figsize=(10, 6))
sns.histplot(data=data['Salário Anual (R$)'],
             bins=20,color='#6495ED',kde=True)

# Adicionando título e rótulos
plt.title('Distribuição de Salários Anuais', fontsize=16)
plt.xlabel('Salário Anual (R$)', fontsize=12)
plt.ylabel('Frequência', fontsize=12)

# Ajustando layout
plt.tight_layout()
plt.show()
```

```python
# Importando os dados
data4 = pd.read_csv('https://raw.githubusercontent.com/netobrandao/Atividades-Alura/refs/heads/main/Praticando%20gr%C3%A1ficos%3A%20distribui%C3%A7%C3%A3o%20com%20Seaborn/atividade_4.csv')

# Construindo o grafico
plt.figure(figsize=(10, 6))
sns.kdeplot(data=data4['Pressão Sistólica (mmHg)'],
            color='#6495ED',
            fill=True,
            alpha=0.6)

# Adicionando título e rótulos
plt.title('Distribuição da Pressão Arterial Sistólica (Assimétrica)', fontsize=16)
plt.xlabel('Pressão Sistólica (mmHg)', fontsize=12)
plt.ylabel('Densidade', fontsize=12)

# Ajustando layout
plt.tight_layout()
plt.show()
```
