# Importando Bibliotecas

```python
#lendo os dados
data1 = pd.read_csv('/home/reginho/Documents/Portifolio Alura/Composição e relacionamento/Atividades dados/atividade_1.csv')
data2 = pd.read_csv('/home/reginho/Documents/Portifolio Alura/Composição e relacionamento/Atividades dados/atividade_2.csv')
data3 = pd.read_csv('/home/reginho/Documents/Portifolio Alura/Composição e relacionamento/Atividades dados/atividade_3.csv')
data4 = pd.read_csv('/home/reginho/Documents/Portifolio Alura/Composição e relacionamento/Atividades dados/atividade_4.csv')
data5 = pd.read_csv('/home/reginho/Documents/Portifolio Alura/Composição e relacionamento/Atividades dados/atividade_5.csv')
data6 = pd.read_csv('/home/reginho/Documents/Portifolio Alura/Composição e relacionamento/Atividades dados/atividade_6.csv')
data7 = pd.read_csv('/home/reginho/Documents/Portifolio Alura/Composição e relacionamento/Atividades dados/atividade_7.csv')
data8 = pd.read_csv('/home/reginho/Documents/Portifolio Alura/Composição e relacionamento/Atividades dados/atividade_8_9.csv')
data9 = pd.read_csv('/home/reginho/Documents/Portifolio Alura/Composição e relacionamento/Atividades dados/atividade_10.csv')
```

# Atividade 1

Visualizando a distribuição do orçamento

```python
#vizualizando os dados
data1.head()
```

```python
# Definindo tamanho o grafico
plt.figure(figsize=(8, 6)) 

# Criando o gráfico de pizza
plt.pie(
    data1['Orcamento (em mil)'],
    labels=data1['Plataforma'],
    autopct='%1.2f%%',
    startangle=135,
    wedgeprops={'edgecolor': 'white'},
    colors=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99'],
    textprops={'fontsize': 12}
)

# Título do gráfico
plt.title('Distribuição do Orçamento de Marketing por Plataforma',fontsize=14, fontweight='bold')

# Exibindo o gráfico
plt.show()
```

# Atividade 2

Verificando as proporções

```python
#vizualizando os dados
data2.head()
```

```python
# Agrupando os dados por categoria e somando as vendas
df_agrupados = data2.groupby('Categoria')['Vendas (em mil)'].sum().reset_index()

# Criando o gráfico de pizza
plt.figure(figsize=(8, 6))
plt.pie(
    df_agrupados['Vendas (em mil)'],
    labels=df_agrupados['Categoria'],
    autopct='%1.1f%%',
    startangle=90,
    colors=['#ff9999', '#66b3ff'],
    wedgeprops={'edgecolor': 'white'},
)

# Título do gráfico
plt.title('Proporção de Vendas por Categoria', fontsize=14, fontweight='bold')

# Exibindo o gráfico
plt.show()
```

# Atividade 3

Acompanhando o impacto financeiro

```python
#vizualizando os dados
data3.head()
```

```python
# Configurando o gráfico de cascata
fig = go.Figure(
    go.Waterfall(
        measure=data3['Tipo'],  # Define o tipo de valor: 'absolute', 'relative', 'total'
        x=data3['Categoria'],  # Categorias no eixo X
        y=data3['Valor'],  # Valores associados a cada categoria
        text=[f'R$ {v:,.2f}' for v in data3['Valor']],  # Formata os rótulos como valores monetários
    )
)

# Adicionando título e personalização
fig.update_layout(
    title="Impacto Financeiro nos Lucros do Restaurante",
    font=dict(size=14),  # Tamanho da fonte
    plot_bgcolor="white",  # Fundo branco para melhor visualização
    yaxis_title="Valores (R$)",  # Rótulo do eixo Y
    waterfallgap=0.5  # Espaçamento entre as barras
)

# Exibindo o gráfico
fig.show()
```

# Atividade 4

Comparando o desempenho por regiões

```python
#vizualizando os dados
data4.head()
```

```python
# Configurando os dados para o gráfico
categorias = ['Alimentos Orgânicos', 'Produtos de Limpeza Sustentáveis', 'Utensílios Ecológicos']
regioes = data4['Região']
valores = [data4[categoria] for categoria in categorias]

# Configurando o eixo X
x = np.arange(len(regioes))

# Criando o gráfico de barras empilhadas
fig, ax = plt.subplots(figsize=(12, 8))
ax.bar(x, valores[0], label='Alimentos Orgânicos', color='#66c2a5')
ax.bar(x, valores[1], bottom=valores[0], label='Produtos de Limpeza Sustentáveis', color='#fc8d62')
ax.bar(x, valores[2], bottom=valores[0] + valores[1], label='Utensílios Ecológicos', color='#8da0cb')

# Configurando os rótulos e título
ax.set_title('Composição do Faturamento por Região e Categoria', fontsize=14)
ax.set_ylabel('Faturamento (R$)', fontsize=12)
ax.set_xticks(x)
ax.set_xticklabels(regioes)
ax.grid(False)
ax.legend(title='Categorias',loc='upper left')

# Exibindo o gráfico
plt.tight_layout()

plt.show()
```

# Atividade 5

Entendendo o relatório das filiais

```python
#vizualizando os dados
data5.head()
```

```python
dados_agrupados = data5.groupby(['Filial', 'Serviço'])['Receita (R$)'].sum().unstack()
dados_agrupados

categorias = dados_agrupados.columns  # Categorias de serviços
filiais = dados_agrupados.index  # Filiais
valores = [dados_agrupados[categoria].values for categoria in categorias]  # Valores por serviço

x = np.arange(len(filiais))

plt.figure(figsize=(10, 6))
plt.bar(x, valores[0], label=categorias[0], color='#1f77b4')
plt.bar(x, valores[1], bottom=valores[0], label=categorias[1], color='#ff7f0e')
plt.bar(x, valores[2], bottom=valores[0] + valores[1], label=categorias[2], color='#2ca02c')


# Configurando rótulos e título

plt.title('Composição das Receitas por Filial e Serviço', fontsize=14)
plt.ylabel('Receita (R$)', fontsize=12)
plt.xticks(x, filiais)  # Define as etiquetas no eixo X
plt.legend(title='Serviços')  # Adiciona a legenda
plt.grid(False)


# Ajustando o layout
plt.tight_layout()
plt.show()
```

# Atividade 6

Desempenho das fontes ao longo do tempo

```python
#vizualizando os dados
data6.head()
```

```python
# Configurando os dados para o gráfico
fig = go.Figure()

# Adicionando as áreas empilhadas para cada fonte de energia
fig.add_trace(go.Scatter(
    x=data6['Data'], y=data6['Solar'],
    mode='lines', fill='tonexty',
    name='Solar',
    line=dict(width=0.5, color='#ff6f61')
))
fig.add_trace(go.Scatter(
    x=data6['Data'], y=data6['Eólica'] + data6['Solar'],
    mode='lines', fill='tonexty',
    name='Eólica',
    line=dict(width=0.5, color='#ffd700')
))
fig.add_trace(go.Scatter(
    x=data6['Data'], y=data6['Hidrelétrica'] + data6['Eólica'] + data6['Solar'],
    mode='lines', fill='tonexty',
    name='Hidrelétrica',
    line=dict(width=0.5, color='#76c7c0')
))

# Configurando layout do gráfico
fig.update_layout(
    title='Composição da Geração de Energia ao Longo do Tempo',
    xaxis_title='Data',
    yaxis_title='Geração de Energia (MWh)',
    legend_title='Fontes de Energia',
    plot_bgcolor='white',
    xaxis=dict(showgrid=True),
    yaxis=dict(showgrid=True),
    font=dict(size=12)
)

# Exibindo o gráfico
fig.show()
```

# Atividade 7

Observando relações entre variáveis

```python
#vizualizando os dados
data7.head()
```

```python
# Configurando o gráfico de dispersão
plt.figure(figsize=(10, 6))
plt.scatter(
    data7['Distância (km)'], 
    data7['Tempo de Entrega (h)'], 
    alpha=0.7, 
    edgecolor='k',
    color='#1f77b4'
)

# Adicionando título e rótulos
plt.title('Relação entre Distância Percorrida e Tempo de Entrega', fontsize=14)
plt.xlabel('Distância Percorrida (km)', fontsize=12)
plt.ylabel('Tempo de Entrega (h)', fontsize=12)

# Ajustando grade para melhor visualização
plt.grid(True, linestyle='--', alpha=0.5)

# Exibindo o gráfico
plt.tight_layout()
plt.show()
```

# Atividade 8 e 9

Avaliando variaveis em vendas com uma reta de tendencia

```python
#vizualizando os dados
data8.head()
```

```python
# Configurando o gráfico de dispersão
plt.figure(figsize=(10, 6))
plt.scatter(
    data8['Tempo no Site (min)'], 
    data8['Valor do Pedido (R$)'], 
    alpha=0.5,
    color='black',
    label='Pedidos'
)

# Adicionando uma linha de tendência
coef = np.polyfit(data8['Tempo no Site (min)'], data8['Valor do Pedido (R$)'], 1)
poly1d_fn = np.poly1d(coef)
plt.plot(
    data8['Tempo no Site (min)'], 
    poly1d_fn(data8['Tempo no Site (min)']), 
    color='#CA0123', 
    linewidth=2, 
    label='Linha de Tendência'
)

# Adicionando título e rótulos
plt.title('Relação entre Tempo no Site e Valor do Pedido com Linha de Tendência', fontsize=14)
plt.xlabel('Tempo no Site (min)', fontsize=12)
plt.ylabel('Valor do Pedido (R$)', fontsize=12)

# Adicionando legenda
plt.legend()

# Ajustando grade para melhor visualização
plt.grid(True, linestyle='--', alpha=0.5)

# Exibindo o gráfico
plt.tight_layout()
plt.show()
```
