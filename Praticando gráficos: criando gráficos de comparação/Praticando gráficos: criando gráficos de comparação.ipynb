```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')
```

# Atividade 1

```python
# Importando os dados e exibindo tabela
data1 = pd.read_csv('https://raw.githubusercontent.com/alura-cursos/python_dados/refs/heads/main/Dados/dados_automoveis.csv')
data1
```

```python
# Agrupando dados
df_tamanho = data1.groupby('Tamanho')['Valor($)'].mean().reset_index()
df_tamanho_ordernado = df_tamanho.sort_values(by=['Valor($)'], ascending=False)
df_tamanho_ordernado
```

```python
# Criando o Gráfico
plt.figure(figsize=(8, 6))
sns.barplot(x='Tamanho', y='Valor($)', data=df_tamanho_ordernado)

# Parametros do Gráfico
plt.title('Valor médio dos automóveis por tamanho', loc='left', fontsize=16)
plt.xlabel('Tamanho', fontsize=12)
plt.ylabel('Valor ($)', fontsize=12)
sns.despine() 

# Exibindo o Gráfico
plt.show()
```

# Atividade 2

```python
# Agrupando dados para serem usados no grafico
df_estilo = data1.groupby(['Estilo'])['Valor($)'].mean().sort_values(ascending=False).reset_index() # Criando tabela que será usada no grafico
df_estilo
```

```python
# Criando o Gráfico
plt.figure(figsize=(8, 6))
sns.barplot(y='Estilo', x='Valor($)', data=df_estilo)

# Parametros do Gráfico
plt.title('Valor médio dos automóveis por estilo', loc='left', fontsize=18)
plt.xlabel('Valor ($)', fontsize=12)
plt.ylabel('')
sns.despine()

# Exibindo o Gráfico
plt.show()
```

# Atividade 5

```python
data3 = pd.read_csv('https://raw.githubusercontent.com/alura-cursos/python_dados/refs/heads/main/Dados/apartamentos_aluguel.csv')
df_regiao = data3.groupby('Regiao')['Valor'].mean().sort_values(ascending=False).reset_index() # Criando tabela que será usada no grafico
df_regiao
```

```python
# Criando o Gráfico
plt.figure(figsize=(10, 6))
sns.barplot(data=df_regiao, x='Regiao',y='Valor',hue='Regiao')

# Parametros do Gráfico
plt.title('Valor médio dos alugueis por região', loc='left', fontsize=16)
plt.xlabel('Região', fontsize=12)
plt.ylabel('Valor do aluguel (R$)', fontsize=12)
sns.despine() 

# Exibindo o Gráfico
plt.show()
```

# Atividade 6

```python
# Importando os dados e exibindo tabela
df_ano = data3.groupby('Ano')['Valor'].mean().sort_values(ascending=False).reset_index() # Criando tabela que será usada no grafico
df_ano
```

```python
# Criando o Grafico
plt.figure(figsize=(10, 6))
sns.lineplot(x='Ano', y='Valor', data=df_ano, linewidth=3)

# Parametros do Gráfico
plt.title('Valor médio dos alugueis por ano de construção do imóvel', loc='left', fontsize=16)
plt.xlabel('Ano', fontsize=12)
plt.ylabel('Valor do aluguel (R$)', fontsize=12)
sns.despine() 

# Exibindo o Gráfico
plt.show()
```

# Atividade 8

```python
# Importando os dados e exibindo tabela
data5 = pd.read_csv('https://raw.githubusercontent.com/alura-cursos/python_dados/refs/heads/main/Dados/dados_jogos.csv')
df_vendas_globais = data5.groupby('genero')['vendas_globais'].sum().sort_values(ascending=False).reset_index() # Criando tabela que será usada no grafico
df_vendas_globais
```

```python
# Criando o Gráfico
plt.figure(figsize=(10, 6))
sns.barplot(data=df_vendas_globais,x='vendas_globais',y='genero')

# Parametros do Gráfico
plt.title('Vendas globais de jogos por gênero', loc='left', fontsize=16)
plt.xlabel('Milhões de dólares ', fontsize=12)
plt.ylabel('', fontsize=12)
sns.despine() 

# Exibindo o Gráfico
plt.show()
```
